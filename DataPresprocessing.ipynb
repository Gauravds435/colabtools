```
import io
```

```
df = pd.read_csv(io.StringIO(uploaded["Car Prices.csv"].decode("utf-8")))
df.head()
```

```
df.columns
```

```
df.drop(['car_ID', 'symboling','fueltype', 'aspiration',
       'doornumber', 'carbody', 'drivewheel', 'enginelocation', 'wheelbase',
       'carlength', 'carwidth', 'carheight', 'curbweight', 'enginetype',
       'cylindernumber','fuelsystem', 'boreratio', 'stroke',
       'compressionratio','citympg', 'highwaympg'], axis = 1, inplace=True)
```

```
df.describe()
```

```
from sklearn.preprocessing import MinMaxScaler
df.head()
```

```
df.drop(["CarName"], axis = 1, inplace = True)
```

```
scaler = MinMaxScaler()
df_values = df.values
df_scaled = scaler.fit_transform(df_values)
normalized_df = pd.DataFrame(df_scaled)
normalized_df.head()
```

```
normalized_df.describe()
```

```
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
df_values = df.values
df_std = std_scaler.fit_transform(df_values)
stand_df = pd.DataFrame(df_std)
stand_df.head()
```

```
stand_df.describe()
```

```
salary = pd.read_csv(io.StringIO(uploade["Engineering_graduate_salary.csv"].decode("utf-8")))
salary.head()
```

```
salary.drop(['DOB', '10percentage', '12graduation',
       '12percentage', 'CollegeID', 'CollegeTier', 'Degree',
       'Specialization', 'collegeGPA', 'CollegeCityID', 'CollegeCityTier',
       'CollegeState', 'GraduationYear', 'English', 'Logical', 'Quant',
       'Domain', 'ComputerProgramming', 'ElectronicsAndSemicon',
       'ComputerScience', 'MechanicalEngg', 'ElectricalEngg', 'TelecomEngg',
       'CivilEngg', 'conscientiousness', 'agreeableness', 'extraversion',
       'nueroticism', 'openess_to_experience', 'Salary'], axis = 1, inplace=True)
```

```
salary.head()
```

```
new = {"f":0, "m":1}
salary = salary.replace({"Gender":new})
salary.info()
```

```
salary
```

```
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
salary["10board"] = lb.fit_transform(salary["10board"])
salary
```

```
salary = pd.get_dummies(salary, columns=["12board"], prefix="12board", prefix_sep=" ")
salary
```
