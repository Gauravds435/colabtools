```
# Combining the values into one
df["TotalBsmtSF"] = df["TotalBsmtSF"] + df["1stFlrSF"] + df["2ndFlrSF"]
df.head()
```

```
df.columns
```

```
# Converting string columns into integer vallues for better prediction
from sklearn.preprocessing import  LabelEncoder
columns = ("Street", "LotShape", "LandSlope", "Condition1", "Condition2", "OverallQual", "ExterQual", "ExterCond","OverallCond",
           "BsmtQual", "BsmtCond", "BsmtExposure","BsmtFinType1", "BsmtFinType2", "HeatingQC", "CentralAir", "KitchenQual",
           "GarageQual", "GarageCond", "SaleCondition", "PavedDrive", "Functional", "GarageFinish", "MSZoning", "LandContour",
           "Utilities", "LotConfig", "Neighborhood", "RoofStyle",	"RoofMatl","Exterior1st",	"Exterior2nd",	"Foundation",
           "Heating", "Electrical", "GarageType", "SaleType")
for c in columns:
  lb = LabelEncoder()
  lb.fit(list(df[c].values))
  df[c] = lb.transform(list(df[c].values))
df.head()
```

```
#  Converting unusual subvalues into desired subvalues 
df["BldgType"]=df["BldgType"].replace("1Fam", "Single Family")
df["BldgType"]=df["BldgType"].replace("2FmCon", "Two-Family Conversion")
df["BldgType"]=df["BldgType"].replace("Duplx", "Duplex")
df["BldgType"] = df["BldgType"].replace("TwnhsE", "Townhouse End")
df["BldgType"] = df["BldgType"].replace("TwnhsI", "Townhouse Inside")
df.head()
```

```
lbl = LabelEncoder()
df["BldgType"] = lbl.fit_transform(df["BldgType"])
df.head()
```

```
#  Converting unusual subvalues into desired subvalues 
df["HouseStyle"] = df["HouseStyle"].replace("1Story", "One story")
df["HouseStyle"] = df["HouseStyle"].replace("1.5Fin", "One and one-half story: 2nd level finished")
df["HouseStyle"] = df["HouseStyle"].replace("2Story", "Two story")
df["HouseStyle"] = df["HouseStyle"].replace("2.5Fin", "Two and one-half story: 2nd level finished")
df["HouseStyle"] = df["HouseStyle"].replace("1.5Unf", "One and one-half story: 2nd level unfinished")
df["HouseStyle"] = df["HouseStyle"].replace("2.5Unf", "Two and one-half story: 2nd level unfinished") 
df["HouseStyle"] = df["HouseStyle"].replace("SFoyer", "Split Foyer")
df["HouseStyle"] = df["HouseStyle"].replace("SLvl", "Split Level")
df.head()
```

```
lbla = LabelEncoder()
df["HouseStyle"] = lbla.fit_transform(df["HouseStyle"])
df.head()
```

```
df.drop(["MasVnrType"], axis=1, inplace = True)
```
